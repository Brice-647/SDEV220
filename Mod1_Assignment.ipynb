```python
# %% [markdown]
# 3.1 How many seconds are in an hour?
# There are 60 seconds in a minute and 60 minutes in an hour.

# %%
60 * 60
```

```python
# %% [markdown]
# 3.2
# Save the answer from the previous question in a variable called seconds_per_hour.

# %%
seconds_per_hour = 60 * 60
seconds_per_hour
```

```python
# %% [markdown]
# 3.3 How many seconds are in a day?
# A day has 24 hours, so we multiply by 24.

# %%
seconds_per_hour * 24
```

```python
# %% [markdown]
# 3.4
# Save the total seconds in a day into a variable called seconds_per_day.

# %%
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

```python
# %% [markdown]
# 3.5 Floating point division
# Divide seconds_per_day by seconds_per_hour using /.

# %%
seconds_per_day / seconds_per_hour
```

```python
# %% [markdown]
# 3.6 Integer division
# Divide seconds_per_day by seconds_per_hour using //.
# This gives the same number as above, just without the .0.

# %%
seconds_per_day // seconds_per_hour
```

```python
# %% [markdown]
# 4.2 Poem using f-strings (instead of old-style formatting)

# %%
food1 = "roast beef"
food2 = "ham"
body_part = "head"
food3 = "clam"

poem = (
    f"My kitty cat likes {food1},\n"
    f"My kitty cat likes {food2},\n"
    f"My kitty cat fell on his {body_part}\n"
    f"And now thinks he's a {food3}."
)

print(poem)
```

```python
# %% [markdown]
# 4.3 Form letter using new-style formatting
# Save the string as 'letter'

# %%
letter = """Dear {salutation} {name},

Thank you for your letter. We are sorry that our {product}
{verbed} in your {room}. Please note that it should never
be used in a {room}, especially near any {animals}.

Send us your receipt and {amount} for shipping and handling.
We will send you another {product} that, in our tests,
is {percent}% less likely to have {verbed}.

Thank you for your support.

Sincerely,
{spokesman}
{job_title}
"""
```
